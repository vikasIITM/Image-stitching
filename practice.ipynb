```python
import numpy as np
import cv2
import glob
import imutils
from google.colab.patches import cv2_imshow
```

```python
img1=cv2.imread("/content/first.jpg",cv2.IMREAD_COLOR)
img2=cv2.imread("/content/first.jpg",cv2.IMREAD_COLOR)
```

```python
cv2_imshow(img1)
```

```python
cv2_imshow(img2)
```

```python
images=[]
images.append(img1)
images.append(img2)
```

```python
stitcher=cv2.Stitcher.create()
ret,pano=stitcher.stitch(images)
```

```python
if ret==cv2.STITCHER_OK:
  cv2_imshow(pano)
  
else:
 print("Images could not be stitched ")
 print("Image does not have enough common keypoints")
```
